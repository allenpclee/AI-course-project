```python
root = 1
cost_list = []
path_list = []
for _ in range(100):
    dv = np.random.randint(1, 16)
    k = np.random.randint(1, 4)
    items = [np.random.randint(1, 64) for _ in range(k) ]

    dv_path, dv_cost = BFS(root, dv)
    it_path, it_cost = IDS(items)
    
    cost_list.append(dv_cost + sum(it_cost))
    path_list.append((dv_path, it_path))
    
    root = dv
```
